# Data Preprocessing and Feature Engineering

```python
import pandas as pd

# Something the file and convert into a dataframe object fro handling
data_path = "Milestone3/milestone3_data/train.csv"
data = pd.read_csv(data_path)
```

```python
#  Check if there are any missing values in the data set
data.isna().values.any()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Separate the features from the target variable
features = data.drop('target', axis=1)

# Center the data by subtracting the mean
centered_data = features - features.mean()

# Normalize the data by dividing by the range (maximum - minimum)
normalized_data = centered_data / (features.max() - features.min())

# Standardize the data using sklearn's StandardScaler
scaler = StandardScaler()
standardized_data = scaler.fit_transform(features)

# Print the first few rows of each transformed dataset
print("Centered data:")
print(centered_data)

print("\nNormalized data:")
print(normalized_data)

print("\nStandardized data:")
print(standardized_data)

```
